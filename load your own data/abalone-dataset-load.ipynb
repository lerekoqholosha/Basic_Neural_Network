## Getting Data

```python
##import packags
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
```

```python
## Load Data 
df=pd.read_csv("/kaggle/input/abalone-dataset/abalone.csv")
```

```python
df.head()
```

The main task we want to do is to predict the age(Rings) using other measurements

Lets split our data

## Split data

```python
df.drop('Sex',inplace=True,axis=1)
```

```python
#features and labels
abalone_features=df.copy()
abalone_labels=abalone_features.pop("Rings")
```

```python
abalone_features
```

For this dataset we will treat all features identically. Pack the features into a single NumPy array.




```python
abalone_features=np.array(abalone_features)
```

```python
abalone_features
```

## Model Building

Next make a regression model predict the age. Since there is only a single input tensor, a tf.keras.Sequential model is sufficient here.

```python
#model
model=tf.keras.Sequential([
    layers.Dense(64),
    layers.Dense(1)]
)

#compile model
model.compile(loss=tf.keras.losses.mean_absolute_error,optimizer=tf.keras.optimizers.Adam())

#fit the model
model.fit(abalone_features,abalone_labels,epochs=10)
```

there we go we have seen how to upload the csv data and build basic model using tensorflow. and our model it seems to be doing well because our loss it's going down(decreasing).but we need test validation to confirm that the. model is doing well.
